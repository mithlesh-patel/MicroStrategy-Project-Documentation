```python
# Program to process project documentation generated by MSTR desktop

# Folders which are not very useful for end users will be skipped from the documentation

# Below are the libraries required to execute this program, please install these if not done so before importing those here



import codecs

# You may need to install BeautifulSoup before importing it into this program, have a look at documentation link above for more information on installation

from bs4 import BeautifulSoup 

import glob

import os
```

```python

# Directory where all html files generated by MSTR Project Documentation wizard are located    

# Replace below highlighted path with the path where you have uploaded documentation folder on web server       

htmlfileLocation = "/local/tmp/ProjectDocumentation/"



# Replace below path with the path where you want to store processed documentation on web server

newhtmlLocation = "/tomcat6/webapps/ProjectDocumentation"



# Replace below path to the path where you have saved all the images generated by documentation wizard on web server

new_image_location = "/ProjectDocumentation/Images/"



# Create a list of GUIDs of the folder which you want to exclude from the documentation  suffixed with .html



hrefs_to_be_removed = ['FF1A02AE11E2C77001730080EF85AD42.html',

                      '84DAAC900B864173907F44B75C68FD96.html']





print('- MSTR project documentation is in progress, the whole process may take several minutes, kindly wait...\n')



# Cleanup destination directory

# Check if any old file generated by any of previous attempts

# If yes then remove those html files from the destination folder



test = newhtmlLocation+'*.html'

r = glob.glob(test)

if len(r) > 0:

    for i in r:

        os.remove(i)

    print('- Old documentation has been removed... \n')

    

# Below variables are to count the number of files being processed by this program

file_count = 0

folderCount = 1

for f in glob.glob(htmlfileLocation+"*.html"):

    html_location_split = f.split("/")

    

    html_length = len(html_location_split)

    html_file_name = html_location_split[html_length-1]

    f=codecs.open(f, 'r', 'latin-1')

    soup= BeautifulSoup(f.read(),"lxml")

    # Below 2 lines will be updating project name with any of the string which you want to use

    # I am simply using 'MicroStrategy Documentation'

    if soup.find("td", {"class": "PROJECTNAME"}).text:

        soup.find("td", {"class": "PROJECTNAME"}).string='MicroStrategy Documentation'

    # Below 2 lines will get the header and update it with any string of your choice

    # I am using 'Object Definition'

    if soup.find("td", {"class": "SUBTITLE"}).text:

        soup.find("td", {"class": "SUBTITLE"}).string='Object Definition'

    # Below loop will check the location of images and update it with the common location which we have declared in previous step

    for img in soup.findAll('img'):

        image_location= img['src']

        image_location_split = image_location.split("\\")

        length = len(image_location_split)

        image_file_name = (image_location_split[length-1]).strip()

        new_file_location = new_image_location+image_file_name

        img['src'] = new_file_location

                 

 

    my_html_string = soup.prettify(formatter=None).encode('latin-1')

    

    # Create a new html file with the same name as source file,

    # update it with processed content and save in destination directory

    

    with open(newhtmlLocation+html_file_name,"wb") as file:

        file.write(my_html_string)

    file.close()

    file_count+=1

    

    # Removed the files which you don't want to include in documentation

    # These are files whose GUIDs are provided in the beginning of the program.

    

if os.listdir(newhtmlLocation) :

    for htFiles in hrefs_to_be_removed:

        if htFiles.endswith(".html"):

            os.remove(newhtmlLocation+htFiles)

            folderCount += folderCount

else:

    print("No older files found in ProjectDocumentation")

    

totalFiles = file_count - (folderCount-1)



print(f"- Microstrategy project documentation has been updated with {totalFiles} files,please verify from MSTR web.")
```
